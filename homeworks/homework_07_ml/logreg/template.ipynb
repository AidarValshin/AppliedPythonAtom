```python
import numpy as np
from logistic_regression import LogisticRegression as  MyLogReg1
```

```python
from sklearn.datasets import make_classification
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score, make_scorer
```

```python
X, y = make_classification(n_samples=1000, n_classes=2, n_informative=14, random_state=43)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
clf = MyLogReg1(lambda_coef=0.4, regulatization="l1", alpha=0.7)
```

```python
clf.fit(X_train, y_train,iterations=1000)
```

```python
clf.predict_proba(X_test)[0][:,np.newaxis]
```

```python
roc_auc_score(y_test, clf.predict_proba(X_test)[0][:,np.newaxis])
```
